# Regression - Yuvraj 

```r
library("tidyverse")
```

```r
FinalData <- read.csv("FinalData.csv")
```

```r
lm.fit <- lm(GRAD_RATE ~ TEACH_STUD_RATIO + FINAL_GRANT, data = FinalData)
```

```r
summary(lm.fit)
```

```r
par(mfrow = c(2, 2))
plot(lm.fit)
```

```r
lm.fit1 <- lm(GRAD_RATE ~ poly(TEACH_STUD_RATIO,2) + poly(FINAL_GRANT,2), data = FinalData)
```

```r
summary(lm.fit1)
```

When we do the non-linear model, we see an improvement in the p-values.

```r
lm.fit1 <- lm(GRAD_RATE ~ poly(TEACH_STUD_RATIO,4) + poly(FINAL_GRANT,3), data = FinalData)
```

```r
summary(lm.fit1)
```

Here, we see that degree 4 and degree 3 of TEACH_STUD_RATIO and FINAL_GRANT, respectively, are significant for this model. Now, we will only include the TEACH_STUD_RATIO^4 and FINAL_GRANT^3 in the model and check against the p-values 

```r
lm.fit1 <- lm(GRAD_RATE ~ I(TEACH_STUD_RATIO^4) + I(FINAL_GRANT^3), data = FinalData)
```

```r
summary(lm.fit1)
```

```r
par(mfrow = c(2, 2))
plot(lm.fit1)
```

Now, we can see a pattern in the residuals. Here, we can say that there is a relationship between (GRAD_RATE) and (TEACH_STUD_RATIO, FINAL_GRANT) 

# Classification -Donovan Unruh

```r
FinalData <- read.csv("FinalData.csv")
```

```r
summary(FinalData)
```

```r
TestData <- FinalData %>% mutate(Qualitative_Grad = if_else(GRAD_RATE >= 93, "High", "Low"))
TestData
```

```r
summary(TestData)
TestData$Qualitative_Grad <- as.factor(TestData$Qualitative_Grad)
summary(TestData)
```

```r
attach(FinalData)
plot(GRAD_RATE, FINAL_GRANT)
plot(GRAD_RATE, TEACH_STUD_RATIO)
```

```r
summary(TestData)
glm.fits <- glm( Qualitative_Grad ~ TEACH_STUD_RATIO + FINAL_GRANT , 
                data = TestData, family = binomial)
summary(glm.fits)
```

```r
## P values show that there is significance
glm.probs <- predict(glm.fits, type = "response")
attach(TestData)
contrasts(Qualitative_Grad)
```

```r
glm.pred <- rep("High", 507)
glm.pred[glm.probs > .5] = "Low"
```

```r
table(glm.pred, Qualitative_Grad)
```

```r
mean(glm.pred == Qualitative_Grad)
```

```r
train <- (SCHOOL_YEAR == "2010/2011")
FinalData.2011 <- FinalData[!train, ]
dim(FinalData.2011)
Grad_2011 <- Qualitative_Grad[!train]
```

```r
glm.fits <- glm(Qualitative_Grad ~ TEACH_STUD_RATIO + FINAL_GRANT,  
                data = TestData, family = binomial, subset = train)
glm.probs <- predict(glm.fits,FinalData.2011,
                    type = "response")
```

```r
glm.pred <- rep("High", 449)
glm.pred[glm.probs > .5] <- "Low"
table(glm.pred, Grad_2011)
mean(glm.pred == Grad_2011) #correct rate
mean(glm.pred != Grad_2011) #error rate
plot(glm.fits)
```

```r
#Trying LDA
```

```r
library(MASS)
```

```r
lda.fit <- lda(Qualitative_Grad ~ FINAL_GRANT + TEACH_STUD_RATIO, subset = train)
lda.fit
plot(lda.fit)
```

```r
lda.pred <- predict(lda.fit,FinalData.2011 )
names(lda.pred)
```

```r
lda.class <- lda.pred$class
table(lda.class, Grad_2011)
mean(lda.class == Grad_2011)
```

```r
sum(lda.pred$posterior[,1] >= .8)
#We do not have a lot of data values in the low percentile
```

# Clustering - Habib Asadullah

```r
FinalData <- read.csv("FinalData.csv")
```

```r
summary(FinalData)
```

```r
attach(FinalData)
plot(GRAD_RATE, TEACH_STUD_RATIO)
plot(GRAD_RATE, FINAL_GRANT)
```

```r
Grant <- FinalData %>% dplyr::select(GRAD_RATE, FINAL_GRANT)
Ratio <- FinalData %>% dplyr::select(GRAD_RATE, TEACH_STUD_RATIO)
```

```r
View(FinalData)
View(Grant)
View(Ratio)
```

```r
wss <- function(k) {
    kmeans(Grant, k, nstart = 10)$tot.withinss
}
```

```r
k.values <- 1:15
wss_values <- map_dbl(k.values, wss)
plot(k.values, wss_values, type = "b", pch = 19, frame = FALSE, xlab="k", ylab="sum_of_squares")
```

```r
kmeans(Grant, 3, nstart = 10)
```

```r
GrantCopy <- FinalData
View(GrantCopy)
```

```r
clusters <- kmeans(Grant, 3, nstart = 10)
```

```r
GrantCopy <- GrantCopy |> mutate(cluster = clusters$cluster)
```

```r
View(GrantCopy)
```

```r
GrantCopy |> ggplot(aes(x = GRAD_RATE, y = FINAL_GRANT, col = as.factor(cluster))) + geom_point()
```

```r
# Now to cluster the teach_stud_ratio
wss <- function(k) {
    kmeans(Ratio, k, nstart = 10)$tot.withinss
}
```

```r
k.values <- 1:15
wss_values <- map_dbl(k.values, wss)
plot(k.values, wss_values, type = "b", pch = 19, frame = FALSE, xlab="k", ylab="sum_of_squares")
```

```r
kmeans(Ratio, 3, nstart = 10)
```

```r
RatioCopy <- FinalData
View(RatioCopy)
```

```r
clusters2 <- kmeans(Ratio, 3, nstart = 10)
```

```r
RatioCopy <- RatioCopy |> mutate(cluster = clusters2$cluster)
```

```r
View(RatioCopy)
```

```r
# now to view them together
RatioCopy |> ggplot(aes(x = GRAD_RATE, y = TEACH_STUD_RATIO, col = as.factor(cluster))) + geom_point()
GrantCopy |> ggplot(aes(x = GRAD_RATE, y = FINAL_GRANT, col = as.factor(cluster))) + geom_point()
```
