```r
TransformedTotalFund <- transform(TotalFund,DISTRICT_NUMBER = as.integer(DISTRICT_NUMBER) ) %>% drop_na()
TransformedTotalFund <- transform(TransformedTotalFund, FINAL_GRANT = as.integer(FINAL_GRANT) ) %>% drop_na()
TransformedTotalFund %>% arrange(School.Year, DISTRICT_NUMBER)
```

```r
FinalG2G<- G2G %>% filter(SUB_POPULATION == "All Students", DATA_LEVEL == "District Level", 
               PUBLIC_OR_INDEPENDENT == "BC Public School", SCHOOL_YEAR != "2001/2002" ) %>% 
        select(SCHOOL_YEAR, DISTRICT_NUMBER, DISTRICT_NAME, STUDENT_GRADE, DENOMINATOR_STUDIED_COHORT, NUMERATOR_NUM_SUCCESSFUL)
        
```

```r
FinalG2G<-transform(FinalG2G, DENOMINATOR_STUDIED_COHORT = as.integer(DENOMINATOR_STUDIED_COHORT)) %>% drop_na()
FinalG2G <- transform(FinalG2G, NUMERATOR_NUM_SUCCESSFUL = as.integer(NUMERATOR_NUM_SUCCESSFUL)) %>% drop_na()
FinalG2G
```

```r
FinalG2G<-FinalG2G %>% group_by(SCHOOL_YEAR, DISTRICT_NUMBER, DISTRICT_NAME) %>% 
            summarise(across(c(DENOMINATOR_STUDIED_COHORT,NUMERATOR_NUM_SUCCESSFUL),sum))
```

```r
FinalG2G
colnames(TransformedTotalFund)[2] = "DISTRICT_NAME"
colnames(TransformedTotalFund)[3] = "SCHOOL_YEAR"
TransformedTotalFund
```

```r
stage1 <- merge(TransformedTotalFund, FinalG2G, by = c("SCHOOL_YEAR", "DISTRICT_NUMBER", "DISTRICT_NAME"))
```

```r
stage1 %>% drop_na()
```

```r
FinalTeachStat<- TeachStat %>%   filter(DATA.LEVEL == "DISTRICT LEVEL"
                    & Educators.CATEGORY == "Teachers" & GENDER == "All",
                SCHOOL.YEAR != "1991/1992", SCHOOL.YEAR != "1992/1993" , SCHOOL.YEAR != "1993/1994",
                SCHOOL.YEAR != "1994/1995", SCHOOL.YEAR != "1995/1996", SCHOOL.YEAR != "1996/1997",
                SCHOOL.YEAR != "1997/1998", SCHOOL.YEAR != "1998/1999", SCHOOL.YEAR != "1999/2000",
                SCHOOL.YEAR != "2000/2001", SCHOOL.YEAR != "2001/2002", SCHOOL.YEAR != "2012/2013",
                SCHOOL.YEAR != "2013/2014" , SCHOOL.YEAR != "2014/2015" , SCHOOL.YEAR != "2015/2016",
                SCHOOL.YEAR != "2016/2017") %>% 
                    select(SCHOOL.YEAR,DISTRICT.NUMBER,Educators.CNT, FULLTIME.Educators.CNT)
```

```r
colnames(FinalTeachStat)[1]= "SCHOOL_YEAR"
colnames(FinalTeachStat)[2]= "DISTRICT_NUMBER"
FinalTeachStat
stage1
Stage2<- merge(stage1, FinalTeachStat , by =c("SCHOOL_YEAR", "DISTRICT_NUMBER"))
```

```r
Stage3 <- transform(Stage2, Educators.CNT = as.integer(Educators.CNT))
Stage4 <- transform(Stage3, FULLTIME.Educators.CNT = as.integer(FULLTIME.Educators.CNT)) %>% drop_na()
Stage5 <- Stage4 %>% mutate(TEACH_STUD_RATIO =  FULLTIME.Educators.CNT/DENOMINATOR_STUDIED_COHORT) %>% arrange(SCHOOL_YEAR,DISTRICT_NUMBER)
```

```r
FinalData <- Stage5  %>% mutate(GRAD_RATE =  NUMERATOR_NUM_SUCCESSFUL/DENOMINATOR_STUDIED_COHORT*100)
```
